```python
from tensorflood.engine import Graph, Variable, Constant, to_graphviz
```

```python
val1, val2, val3 = 2, 10, 7

with Graph() as g:
    x = Variable(val1, name='x')
    y = Variable(val2, name='y')
    c = Constant(val3, name='c')
    
    z = (x * y) + c
    print("Result:", z.data)
    
    z.backward()
    
    # Check
    print("z:", z.data == (val1 * val2) + val3)    
    print("dz/dx:", x.grad == val2)
    print("dz/dy:", y.grad == val1)
    print("dz/dc:", c.grad == 1)
    
to_graphviz(z)
```
